```
import numpy as np
import matplotlib.pyplot as plt

# =====================================================
# PART 1: SciPy Derivative
# =====================================================

# Try the new SciPy derivative first.
# If unavailable, use a numerical central difference.
try:
    from scipy.differentiate import derivative

    def equation(x):
        return x**2 + 3*x + 2

    x0 = 2

    result = derivative(equation, x0)

    print("===================================")
    print("SCIPY DERIVATIVE")
    print("===================================")
    print("Equation: f(x) = x^2 + 3x + 2")
    print("Derivative at x =", x0)
    print(result.df)

except Exception:

    print("===================================")
    print("SCIPY DERIVATIVE (Fallback)")
    print("===================================")

    def equation(x):
        return x**2 + 3*x + 2

    def derivative(func, x, h=1e-6):
        return (func(x+h)-func(x-h))/(2*h)

    x0 = 2

    print("Equation: f(x)=x^2+3x+2")
    print("Derivative at x =", x0)
    print(derivative(equation, x0))

# =====================================================
# PART 2: Gradient Descent
# =====================================================

print("\n===================================")
print("GRADIENT DESCENT")
print("===================================\n")

# Dataset

X = np.array([
    [1, 3],
    [4, 10]
], dtype=float)

y = np.array([
    [5],
    [6]
], dtype=float)

# Initial parameters

m = np.array([
    [-1.0],
    [2.0]
])

b = 1.0

learning_rate = 0.01
iterations = 3

# History for plots

m1_history = []
m2_history = []
b_history = []
error_history = []

# =====================================================
# Gradient Descent Loop
# =====================================================

for i in range(iterations):

    print("===================================")
    print("Iteration", i+1)
    print("===================================")

    # -------------------------------
    # Prediction
    # -------------------------------

    print("\nStep 1: Prediction")

    y_pred = X @ m + b

    print(y_pred)

    # -------------------------------
    # Error
    # -------------------------------

    print("\nStep 2: Error")

    error = y_pred - y

    print(error)

    # -------------------------------
    # Mean Squared Error
    # -------------------------------

    mse = np.mean(error**2)

    print("\nStep 3: Mean Squared Error")

    print(mse)

    # -------------------------------
    # Gradient of m
    # -------------------------------

    print("\nStep 4: Gradient of m")

    grad_m = (2/len(y)) * (X.T @ error)

    print(grad_m)

    # -------------------------------
    # Gradient of b
    # -------------------------------

    print("\nStep 5: Gradient of b")

    grad_b = (2/len(y)) * np.sum(error)

    print(grad_b)

    # -------------------------------
    # Update m
    # -------------------------------

    print("\nStep 6: Update m")

    print("Old m")

    print(m)

    m = m - learning_rate * grad_m

    print("New m")

    print(m)

    # -------------------------------
    # Update b
    # -------------------------------

    print("\nStep 7: Update b")

    print("Old b =", b)

    b = b - learning_rate * grad_b

    print("New b =", b)

    # -------------------------------
    # Store history
    # -------------------------------

    m1_history.append(m[0,0])
    m2_history.append(m[1,0])
    b_history.append(b)
    error_history.append(mse)

# =====================================================
# Final Prediction
# =====================================================

print("\n===================================")
print("FINAL PARAMETERS")
print("===================================")

print("\nFinal m")

print(m)

print("\nFinal b")

print(b)

final_prediction = X @ m + b

print("\nPrediction using final parameters")

print(final_prediction)

print("\nActual y")

print(y)

# =====================================================
# Plot 1
# =====================================================

plt.figure(figsize=(8,5))

plt.plot(
    range(1, iterations+1),
    m1_history,
    marker='o',
    label='m1'
)

plt.plot(
    range(1, iterations+1),
    m2_history,
    marker='s',
    label='m2'
)

plt.plot(
    range(1, iterations+1),
    b_history,
    marker='^',
    label='b'
)

plt.title("m and b over Iterations")

plt.xlabel("Iteration")

plt.ylabel("Parameter Value")

plt.xticks(range(1, iterations+1))

plt.grid(True)

plt.legend()

plt.show()

# =====================================================
# Plot 2
# =====================================================

plt.figure(figsize=(8,5))

plt.plot(
    range(1, iterations+1),
    error_history,
    marker='o'
)

plt.title("Mean Squared Error over Iterations")

plt.xlabel("Iteration")

plt.ylabel("MSE")

plt.xticks(range(1, iterations+1))

plt.grid(True)

plt.show()
```
